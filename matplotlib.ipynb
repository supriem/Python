```python
import numpy as np
t=np.arange(0.,5.,0.2) #means value betn 0-5 with height 0.2

plt.plot(t,t**2,'b--',label='^2')  #blue dashed line
plt.plot(t,t**2.2, 'rs', label='^2.2')  #rs=red square 
plt.plot(t,t**2.5, 'g^',label='^2.5')   # means green triangle
         
plt.grid()
plt.legend()  #notation description os called legend
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y,linewidth=5.0)
plt.show()



```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]

lines=plt.plot(x1,y1,x2,y2) #this plot 2 lines

plt.setp(lines[0], color='r',linewidth=2.0)  #set property =setp
plt.grid()

plt.setp(lines[1],color='b',linewidth=2.0)
```

```python
#for multiple plots
def f(t):  #function
    return np.exp(-t)*np.cos(2*np.pi*t) 

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure(1)  #creates figure

plt.subplot(211) #means subplot with 2 rows and 1 columns, figure no.
plt.grid()
plt.plot(t1,f(t1),'b-')

plt.subplot(321)

plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
plt.figure(1)       #1st figure
plt.subplot(211)    #1st subplot of fig1
plt.plot([1,2,3])
plt.subplot(212)     #second subplot
plt.plot([4,5,6])

plt.figure(2)       #a second figure
plt.plot([4,5,6])   #creates subplot(111) by default

#two plots will be creates
#After this if we want to go in fig 1 then call fig1

plt.figure(1)
plt.figure(211)          #i want to write title in this subplot 
plt.title('Easy sa 1,2,3')
plt.show()
```
