# Displaying the image in Ubuntu terminal

1>We use ANSI sequence to display the image in the terminal of the image.This is an standard which explains what should be the format of the text.

```python
print("\x1b[48;5;4mHey,there!\x1b[0m") #this will print the hey there with blue background#4=blue 2=green

```

we can define 256 color palette in linux terminal.

```python
def get_ansi_color_code(r,g,b):  #this function converts rgb value to 256 bit Ansi standard
    if r==g and g==b:
        if r<8:
            
            return 16
        
        if r>248:
            
            return 231
        return round(((r-8)/247)*24)+232
    return 16+(36*round(r/255*5))+(6*round(g/255*5))+round(b/255*5)


    
```

```python
def get_color(r,g,b):
    return f"\x1b[48;5;{int(get_ansi_color_code(r,g,b))}m \x1b[0m"
```

```python
print(get_color(255,0,0))#passing values to function.Since we passed value 255 to r so this will shows red color
```

To load image we need PIL library.So "pip install pillow"will install pillow library.This library is used for image manipulation.
We also need Numpy.

```python
from PIL import Image
import numpy as np
```

```python
img=Image.open("2.jpeg")#place image in same folder whereyour ipynb file is
```

we are resizing image

```python
h=100
w=int((img.width/img.height)*h)
```

```python
img=img.resize((w,h),Image.ANTIALIAS)#antialisa will not show broken edges
```

```python
img.size #now image has resized to 100*100
```

```python
img_arr=np.asarray(img)
```

```python
img_arr.shape #3 means each pixel has rgb value
```

```python
for x in range(h):
    for y in range(w):
        pix=img_arr[x][y]
        #pix[0],[1],[2] is passing value to function.end=this will not print new line after one pixelhas completed
        print(get_color(pix[0],pix[1],pix[2]),end='',sep='')
    print('2.jpeg')
           
              
              
```
